```python
import numpy as np
import os
```

```python
file_path = '../data/6mwm_pdbs/6mwm_silent_1000_tries.out'

file_name = os.path.splitext(os.path.basename(file_path))[0]

print(file_name)
```

```python
# Open the input file
with open(file_path, 'r') as file:
    lines = file.readlines()

# Extract the names from each line and store them in a list
names = []
first = True
for line in lines:
    if line.startswith("SCORE:"):
      if first == False :
        parts = line.split()
        name = parts[-1]
        names.append(name)
      else: first = False

# Write the names to a file
output_file_path = f'../data/6mwm_pdbs/{file_name}.txt'

with open(output_file_path, 'w') as file:
    for name in names:
        file.write(name + '\n')
```

```python
with open(output_file_path, 'r') as file:
    contents = file.read()

print(contents)
```
